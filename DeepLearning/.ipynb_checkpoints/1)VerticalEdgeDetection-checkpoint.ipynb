## Convolving the image, given image and filter

```python
def convolve2d(image,kernel):
    # defining the padded image
    img_padded = np.zeros((image.shape[0]+kernel.shape[0]-1,image.shape[1]+kernel.shape[1]-1))
    
    # defining the output image
    output = np.zeros(image.shape)
    
    # boundary for copying the original image to padded image
    l_w = int(kernel.shape[0]/2)
    h_w = img_padded.shape[0]-int(kernel.shape[0]/2)
    l_h = int(kernel.shape[1]/2)
    h_h = img_padded.shape[1]-int(kernel.shape[1]/2)
    
    # copying original image to padded image
    img_padded[l_w:h_w,l_h:h_h] = image
    
    # convolving the image
    for i in range(image.shape[0]-kernel.shape[0]):
        for j in range(image.shape[1]-kernel.shape[1]):
            #print(str(i)+" "+str(j)+" "+str(image.shape[1]))
            output[i][j] = np.sum(kernel*img_padded[i:i+kernel.shape[0],j:j+kernel.shape[1]])
    # returning the normalized image
    return normalize(output)    

print(s)
```
