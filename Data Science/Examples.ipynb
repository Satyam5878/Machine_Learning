#### ECDF Plotting( Empirical cumulative distribution plot)

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
# ECDF plot
x = np.sort(height)
y = np.arange(1,len(x)+1)/len(x)
plt.plot(x,y,marker='.',linestyle="none")
plt.grid()
plt.show()
```

```python
x2 = np.sort(height)[:10]
y2 = np.arange(1,10+1)/10
plt.plot(x2,y2,marker='.',linestyle="none")
plt.grid()
plt.show()
```

```python
# Bee Swarm Plot
category = np.random.choice(np.arange(1,5),5000)
df = pd.DataFrame(np.column_stack((height,category)),columns=["Height","Category"])
sns.swarmplot(x="Category",y="Height",data=df)
plt.show()
```
