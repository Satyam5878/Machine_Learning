```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import datetime

print("Done")
```

```python
path  = "C:\\Users\\Aman\\Desktop\\OtherFolders\\Sem7\\Topics Of Study\\Python\\Matplotlib"
file_name = "finance.txt"
file_new_name = "finance.csv"
```

```python
## reading file and converting date to unix timestamp and int csv format
with open(path+"\\finance.csv",'w') as file_new:
    with open(path+"\\"+file_name,'r') as file:
        line = file.readline()
        line_split = line.split(" ")
        for line in line_split:
            line_attr = line.split(',')
            #code for converting date
            line_attr[0] = datetime.datetime.strptime(line_attr[0],"%Y-%m-%d").timestamp()
            #merging the attributes
            line = ''.join(str(item)+',' for item in line_attr)
            #print(line[:-1])
            file_new.write(line[:-1]+"\n")
print("Done")
```

```python
finance = pd.read_csv(path+"\\"+file_new_name)
finance.columns = ["Date","Open","High","Low","Close","Adjusted_close","Volume"]
#finance.scatter(finance["Date"],finance["Close"])
#print(finance.head())
#finance.plot(finance['Date'],finance["Close"])
#finance.plot()
plt.scatter(finance["Date"],finance["Close"] ,linestyle='-')
plt.show()
```
