```python
import matplotlib.pyplot as plt
import random
#

fig = plt.figure()
ax1 = plt.subplot2grid((1,1),(0,0))

x = [x for x in range(0,100)]
y = [int(random.random()*100+1) for _ in range(0,100)]

ax1.plot(x,y)
for label in ax1.xaxis.get_ticklabels():
    label.set_rotation(45)
    
ax1.grid(True)
plt.show()
```
