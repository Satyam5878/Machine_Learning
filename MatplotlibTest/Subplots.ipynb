```python
### Date 05-mar-2018 while tensorflow

import matplotlib.pyplot as plt
import numpy as np

# damping function
def fun(t):
    return np.exp(-t)*np.cos(2*np.pi*t)

t1 = np.arange(0,5,.1)
t2 = np.arange(0,5,.02)

plt.figure()
plt.subplot(221)
plt.plot(t1,fun(t1))


plt.subplot(222)
plt.plot(t2,np.cos(2*np.pi*t2))


plt.subplot(223)
plt.plot(t1,fun(t1))

plt.show()
```
