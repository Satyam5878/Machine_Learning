```python
import matplotlib.pyplot as plt
import numpy as np
#import matplotlib._color_data as mcd
print("Done")
```

```python
#plotting the sin curve
x = np.linspace(0,2*np.pi,100)
y = np.sin(x)

## for grid control
xticks = np.linspace(0,7,15)
yticks = np.linspace(-1,1,11)

plt.plot(x,y ,label='Sine curve')

plt.xlabel('Angle in radians')
plt.ylabel('Magnitude')

plt.title('Sine Curve')
plt.legend()
plt.grid()

plt.xticks(xticks)
plt.yticks(yticks)

plt.axis([0,6.5,-1,1])
plt.show()
```

```python
## plotting multiple curve in the same figure

# using superposition

# data
x = np.linspace(0,2*np.pi,100)
y = np.sin(x)
z = np.cos(x)

plt.plot(x,y,'r',x,z,'g')

# or

plt.plot(x,y,'r')
plt.plot(x,z,'g')

plt.xlabel('Angle in radians')
plt.ylabel('Magnitude')

plt.title('Sine and Cosine Curve')
plt.legend(['Sine','Cosine'])
plt.grid()

plt.xticks(xticks)
plt.yticks(yticks)

#plt.axis([0,6.5,-1,1])
plt.show()

```

```python
## plotting two curve with same x axis but different y axis range

x = np.linspace(0,2*np.pi,100)
y = np.sin(x)
z = np.sinh(x)

xticks = np.linspace(0,7,15)
yticks1 = np.linspace(-1,1,11)
yticks2 = np.linspace(0,300,15)

fig,ax1 = plt.subplots()
ax2 = ax1.twinx()       # creating axes with common x-axis and different y-axis

curve1, =  ax1.plot(x,y,label="Sine curve",color='r')
curve2, = ax2.plot(x,z,label="Sinh curve",color='b')

## legends
curves = [curve1,curve2]
ax1.legend(curves,[curve.get_label() for curve in curves])
ax2.legend([curve1,curve2],["Sine curve2","Sinh curve2"])
## legends
#ax1.legend()
#ax2.legend()

## setting x-label
ax1.set_xlabel("Angle/Value",color=curve1.get_color())
#ax2.set_xlabel("Angle/Value",color=curve2.get_color())   # this won't work because ax2 does not have property over x 
                                                          # w.r.t label,grid,color
## setting y-label
ax1.set_ylabel("Magnitude",color=curve1.get_color())
ax2.set_ylabel("Magnitude",color=curve2.get_color())

## coloring ticks
ax1.tick_params(axis='y',color=curve1.get_color())
ax2.tick_params(axis='y',color=curve2.get_color())

## setting ticks
ax1.set_xticks(xticks)
ax1.set_yticks(yticks1)
ax2.set_yticks(yticks2)

# alternate appraoch 
ax1.xaxis.label.set_color(curve1.get_color())


# grid   
#ax1.grid(color=curve1.get_color())            # grid corresponding to curve 1

# grid corresponding to second axis  i.e curve2
ax1.grid(color=curve2.get_color())
ax2.grid(color=curve2.get_color())
ax1.yaxis.grid(False)

plt.title("Plotting two curves")
plt.show()
```

```python
## plotting two curve with same y-axis and different x-axis
y = np.linspace(0,2*np.pi,100)
x1 = np.sin(y)
x2 = np.sinh(y)

fig,ax1 = plt.subplots()

ax2 = ax1.twiny()

curve1, = ax1.plot(x1,y, label='sin',color='r')
curve2, = ax2.plot(x2,y,label='sinh',color='b')

curves = [curve1,curve2]
#ax1.legend(curves,[curve.get_label() for curve in curves])
ax1.legend(curves,[curve.get_label() for curve in curves])

ax1.set_xlabel("Magnitude",color=curve1.get_color())
ax2.set_xlabel("Magnitude",color=curve2.get_color())
ax1.set_ylabel("Angles/value",color=curve1.get_color())
#ax2.set_ylabel("Angles/value",color=curve2.get_color())

ax1.tick_params(axis='y',color=curve1.get_color())
ax1.tick_params(axis='x',color=curve1.get_color())
ax2.tick_params(axis='x',color=curve2.get_color())


# grid along curve 1
#ax1.grid(color=curve1.get_color())

# for grid along curve 2
ax1.grid(color=curve2.get_color())
ax2.grid(color=curve2.get_color())
ax1.xaxis.grid(False)
ax1.yaxis.grid(True)

plt.title("\n\nCurve with different y-axis")
plt.show()
```

```python
## Moving x-axis and y-axis

x = np.linspace(-2*np.pi,2*np.pi,201)
y1 = np.sin(x)
y2 = np.cos(x)

# drawing graph
plt.plot(x,y1,label='Sine',color='blue')
plt.plot(x,y2,label='Cosine',color='red')

# shiftng x-axis
plt.axes().spines['bottom'].set_position(('data',0))

# shifting y-axis
plt.axes().spines['left'].set_position(('data',0))


# putting horizontal line below
#plt.axhline(y=-1)
plt.axhline(y=plt.ylim()[0],color='k')

# putting vertical line on left position
#plt.axvline(x=-7)
plt.axvline(x=plt.xlim()[0],color='k')

plt.xlabel("Angle")
plt.ylabel("Magnitude")

plt.legend()
plt.show()

```

```python
## histogram

mean = 100
sd = 15
N = 1000
binsize = 20

data = np.random.normal(mean,sd,N)

#plt.hist(data,binsize,facecolor='b',label='histogram',normed=False)
count,bins,extras = plt.hist(data,binsize,facecolor='b',label='Histogram',normed=False)

plt.plot(bins[1:],count,color='red')      # plotting bins and count

print(len(bins))                          # length of bins
print(len(count))                         # lenght of count
plt.xlabel("IQ")
plt.ylabel("Count")
plt.legend()
#plt.xticks(bins)
plt.title("IQ Distribution")
plt.show()
```
