```python
import matplotlib.pyplot as plt
import csv 
#Loading Data from file
x = []
y = []

with open("TestData",'r') as file:
    rows = csv.reader(file,delimiter=',')
    for row in rows:
        x.append(int(row[0]))
        y.append(int(row[1]))

        
plt.plot(x,y,label="Loaded from file")
plt.show()
                 
```

```python
import matplotlib.pyplot as plt
import numpy as np 
#Loading Data from file
#using numpy

x,y = np.loadtxt('TestData',delimiter=",",unpack=True)
plt.plot(x,y,label="Loaded from file")
plt.show()
```
