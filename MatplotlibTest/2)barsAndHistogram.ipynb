```python
import matplotlib.pyplot as plt

#bars 
x = [x for x in range(0,10 ,2)]
y = [6,7,8,2,4]
x1 = [x for x in range(1,11,2)]
y1 = [7,8,2,4,2]

#print(x1)
plt.bar(x,y ,label="Bar1",color="r")

plt.bar(x1,y1,label="Bar2",color="c")
plt.show()
```
