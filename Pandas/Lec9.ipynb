```python
df = pd.read_pickle(path+"\\FMAC_HPI.pickle")
df.set_index("Date",inplace = True)
print(df.head())
```

```python
## converting to Index to dateTime Index
print(type(df.index))
df.index = pd.to_datetime(df.index)
print(type(df.index))
```

```python
## resampling anually and using mean
df_tx = df['TX'].resample("A").mean()
print(df_tx.head())
```

```python
## resampling anually using ohlc
df_tx_ohlc = df['TX'].resample("A").ohlc()
print(df_tx_ohlc.head())
```

```python
## plotting above two 
df['TX'].plot()
df_tx.plot()
df_tx_ohlc.plot()
plt.show()
```

```python
## upsampling
df_tx_down = df["TX"].resample("D").mean()
df_tx_down.plot()
plt.show()
```
